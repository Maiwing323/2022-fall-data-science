## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.

```python
# 1. How many listings are there with a price less than 100? 
select_condiction = df['price'] < 100
df_priceLess = df[select_condiction]

# count() does count the null value
# be ware to use to count()
# df_priceLess['price'].count()
# len function also got the same answer if there is no null value when using count function
#shape function give the value of num of row * num of column
len(df_priceLess)



```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.

condiction_bk = df['neighbourhood_group'] == 'Brooklyn'
df_bk = df[condiction_bk]
df_bk.head()
len(df_bk['neighbourhood_group'])


# delect column using del 
# del df['select_condiction']
# del df['is_in_bk']
# df.head()

```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.
condiction_price = df_bk['price'] < 100
bk_low = df_bk[condiction_price]
len(bk_low)

```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
list_of_hoods = ['Michael','John', 'Daniel']
select_statement = df.host_name.isin(list_of_hoods)
df[select_statement]

```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

con1 = df['neighbourhood'] == 'Williamsburg'
df['adjusted_price'] = np.where(con1, df.price + 100, df.price)
df
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared. 
total_rooms = len(df)  # Total number of rooms
private_rooms = len(df[df['room_type'] == 'Private room'])  # Count of private rooms
shared_rooms = len(df[df['room_type'] == 'Shared room'])  # Count of shared rooms

percentage_private = (private_rooms / total_rooms) * 100
percentage_shared = (shared_rooms / total_rooms) * 100
print("percentage_private: ", percentage_private,"percentage_shared: " ,percentage_shared )


```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

neighbourhood_group = df.groupby('neighbourhood_group')
neighbourhood_group['id'].count()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
neighbourhood_group['price'].mean()

```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 

g3 = df.groupby('neighbourhood_group')
g3['price'].agg(['min', 'max'])

```

```python
g3 = df.groupby('neighbourhood_group')
g3['price'].agg(['min', 'max']).reset_index()
```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.
double_gb = ['neighbourhood_group', 'room_type']
g4 = df.groupby(double_gb)
g4['price'].mean().reset_index()

```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.

double_gbm = ['neighbourhood_group', 'room_type']
g5 = df.groupby(double_gbm)
g5['price'].agg(['min', 'max', 'mean', 'median', 'std']).reset_index()


```
